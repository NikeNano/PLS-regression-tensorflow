```python
from sklearn.cross_decomposition import PLSRegression
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
X = iris.data  # we only take the first two features.
y = iris.target
```

```python
pls = PLSRegression(n_components=2)
```

```python
pls.fit(X,y)
```

```python
pls.x_weights_
```

```python
pls.transform(X)
```
